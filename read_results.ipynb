```python
import glob
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
paths = glob.glob("**/*.log", recursive=True)
```

```python
results = []

for path in paths:
    with open(path, "r") as f:
        res = f.read()
    tmp = re.findall("Accuracy ([0-9\.]+)", res)
    results.append([float(x) for x in tmp])
```

```python
print(f"Experiments: {len(results)}")
print(f"Mean final training accuracy: {np.mean([x[-1] for x in results])}")
print(f"Min final training accuracy: {min([x[-1] for x in results])}")
print(f"Max final training accuracy: {max([x[-1] for x in results])}")
print(f"Min initial training accuracy: {min([x[0] for x in results])}")
print(f"Max initial training accuracy: {max([x[0] for x in results])}")
```

```python
fig, ax = plt.subplots(1, 1)

idxs = np.random.choice(len(results), size=10)
for i, res in enumerate(results):
    if i in idxs:
        ax.plot(res)
    
# ax.set_yticks(np.linspace(min([min(x) for x in results]), max([max(x) for x in results]), num=9))
```
